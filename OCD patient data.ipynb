
# OCD Patient data: Demographic and clinical data analysis
## Objective
The aim of this project is to perform an exploratory data analysis (EDA) on a dataset containing demographic and clinical data of OCD patients. The analysis will focus on understanding the relationships between various demographic factors and clinical outcomes.


### Dataset overview
#### The columns in the dataset
* Patient ID: Unique identifier for each patient.
* Age: Age of the patient.
* Gender:Gender of the patient.
* Ethnicity: Ethnicity of the patient.
* Marital Status: Marital status of the patient.
* Educational Level: Level of education attained by the patient.
* OCD Diagnosis Date: Date when OCD was diagnised.
* Duration of Symptoms (months): Duration for which the patient has been experiencing symptoms
* Previous Diagnoses: Any previous Diagnosis before OCD
* Family History of OCD: Wheather the patient has the family history of OCD
* Obsession Type: Type of obsession experienced by the patient
* Compulsion Type: Type of compulsions experienced by the patient
* Y-BOCS Score (Obsessions): Y-BOCS Score related to obsession
* Y-BOCS Score (Compulsions): Y-BOCS Score related to compulsion
* Depression Diagnosis: Wheather the patient has diagnosied with Depression
* Anxiety Diagnosis: Wheather the patient has diagnosied with Anxiety
* Medications: Medications the patient is currently taking



#### Importing the required libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
```

#### Loading the dataset

```python
df = pd.read_csv('ocd_patient_dataset.csv')
```

```python
# The below command displays the first five rows of the dataset to get a transparent view on data and it helps us to  differentiante 
# how many qualitative and quantitavite data columns were present
df.head()
```

#### Cleaning the data

```python
# Removing the duplicates from the data 
df = df.drop_duplicates()
df
```

```python
# Removing the Nan values from the data 
df = df.fillna('')
df
```

#### Statastical Analysis of data

```python
# The below command gives the information regarding statastical factors with integer data type
df.describe()
```

```python
# The below command gives the information regarding statastical factors with object data atype
df.describe(include=['O'])
```

#### Visualization of data 

```python
# Age Distribution of data using histplot
sns.histplot(df['Age'],bins=30,kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('frequency')
plt.show()
```

```python
# Gender distribution of data using countplot
sns.countplot(x='Gender',data = df)
plt.title('Gender Distrubution of Patients')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()

```

```python
# Ethnicity distrubution of patients using countplot
sns.countplot(x='Ethnicity',data=df)
plt.title('Ethnicty Distrubution of Patients')
plt.xlabel('Ethnicity')
plt.ylabel('count')
plt.show()
```

```python
# Distrubution of Duration of symptoms using histplot
sns.histplot(x='Duration of Symptoms (months)',data = df,bins=20,kde=True)
plt.title('Distrubution of Duration of symptoms')
plt.xlabel('Duration(months)')
plt.ylabel('Frequency')
plt.show()
```

```python
#Comparing the Y-Bocs Scores(obsessions) between male and female using boxplot
sns.boxplot(x='Gender',y='Y-BOCS Score (Obsessions)',data = df,color='orange')
plt.title('Y-Bocs obssesion score by Gender')
plt.xlabel('Gender')
plt.ylabel('Y-BOCS Score (Obsessions)')
plt.show()
```

```python
#  extracting the relationship between obsession and compulsion Y-BOCS scores using lm plot
sns.lmplot(x='Y-BOCS Score (Obsessions)',y='Y-BOCS Score (Compulsions)',data=df)
plt.title('Relationship between Y-BOCS scores (obsession vs compulsion)')
plt.xlabel('Y-BOCS Score (Obsessions)')
plt.ylabel('Y-BOCS Score (Compulsions)')
plt.show()
```

```python
# extracting the relation between Medications used depending on Duration of symptoms(months)
sns.barplot(x='Medications',y='Duration of Symptoms (months)',data = df )
plt.title('Relationship between Duration of Symptoms(months),Medications')
plt.xlabel('Medications')
plt.ylabel('Duration of Symptoms (months)')
plt.show()
```

```python
# identifying the correlation between age, duration of symptoms, Y-BOCS Score(obsessions and compulsions)
corr_matrix = df[['Age','Duration of Symptoms (months)','Y-BOCS Score (Obsessions)','Y-BOCS Score (Compulsions)']].corr()
sns.heatmap(corr_matrix,annot=True,cmap='coolwarm')
plt.title('Corelation Matrix')
plt.show()
```

#### Key insights and findings from statstical analysis and data visualizations

* Male patients are higher in count than female
* Based on Age patients between sixty to seventyfive are more
* Female patients have high Y-Bocs(obsession) score compared with men which means that female patients have more severity
* fifty percent of patients have family history of OCD which means the probability of occurence through genetical
* Nearly fifty percent of patients have either Depression diagnosis or Anxiety diagnosis or both
* Based on severity Durations of Symptoms(months) the medications were changed. The patients having high duration of Symptoms(months) were medicated with SNRI and people with moderate level were treated with Benzodiazepide and patients with low level severity were medicated with SSRI  

#### Output and Interpretation

* After the statastical analysis and data visualization of data The OCD is not correlated with any demographic and clinical factors mostly and they are independently correlated with each other based on Gender and Age
* There is a probability of occurence of OCD with family history ,depression diagnosis and Anxiety diagnosis as fifty percent of patients are of these catogery.
* The severity seems to be more in females than males even though men are affected more with ocd
* The medications were changed depending on the durtion of symptoms i.e, depending on severity.
